```python
#convertCols
# function plot cols-nan
def plotNaN(df, figsize=(10,10)):
    fig = plt.figure(figsize=figsize)
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    ax = fig.add_subplot(2, 2, 1)
    df.isna().sum().sort_values(ascending=False).plot(kind='pie',); # missing values # subplots=(1,1)
    ax = fig.add_subplot(2, 2, (2,2))
    ( (df.isnull().sum() / df.shape[0]) * 100 ).sort_values().plot(kind='barh'); # subplots=(1,2)
    plt.title('Bar Graph of Mission Values')
    
    ax = fig.add_subplot(2, 1, 2)
    ( (df.isnull().sum() / df.shape[0]) * 100 ).sort_values().plot(kind='barh')
    plt.title('WHICH IS THIS')

    
plotNaN(df, (20, 15))
```
